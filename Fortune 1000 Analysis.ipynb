```python
import pandas as pd
import os
import numpy as np
import plotly_express as px
```

```python
df = pd.read_csv(r'C:\Users\Damon\OneDrive\Documents\Personal Portfolio\Fortune 1000 companies\Fortune 1000 Companies by Revenue.csv')
df
```

```python
df.dtypes
```

```python
df.columns
```

```python
#Convert revenues to float
df['revenues '] = df['revenues '].str[1:]
df['revenues '] = df['revenues '].str.replace(',','')
df['revenues '] = df['revenues '].astype(float)
df
```

```python
#Convert profits  to float
df['profits '] = df['profits '].str.replace('$','')
df['profits '] = df['profits '].str.replace('(','-')
df['profits '] = df['profits '].str.replace(')','')
df['profits '] = df['profits '].str.replace(',','')
df['profits '] = df['profits '].str.strip()
df['profits '] = pd.to_numeric(df['profits '], errors= 'coerce')
```

```python
#Convert assets to float
df['assets'] = df['assets'].str[1:]
df['assets'] = df['assets'].str.replace(',','')
df['assets'] = df['assets'].astype(float)
```

```python
df
```

```python
#Convert market_value  to float
df['market_value '] = df['market_value '].str[1:]
df['market_value '] = df['market_value '].str.replace(',','')
df['market_value '] = pd.to_numeric(df['market_value '], errors='coerce')
```

```python
df
```

```python
#Convert employeesto integer
df['employees '] = df['employees '].str.replace(',','')
df['employees '] = pd.to_numeric(df['employees '], errors='coerce')
```

```python
df.dtypes
```

```python
px.scatter(df,'revenues ','profits ',color='name ', trendline='rolling', title='Revenues vs. Profits by Company')
```

```python
df['Profit Margin'] = df['profits '] / df['revenues ']
```

```python
top5 = df.groupby('name ').sum().sort_values(by='Profit Margin', ascending=False).head(10).reset_index()
top5
```

```python
px.bar(top5,'name ','Profit Margin', color='profits ', title="Top 10 'Fortune 1000' companies with highest profitability", labels=True)
```
